Notebook to perform policy iteration on Frozen Lake environment

```python
import gym
import matplotlib.pyplot as plt

from utils import pol_iter, find_opt_q, fancy_visual
```

```python
# Initialize gym environment
env = gym.make('FrozenLake-v1', desc=None, map_name="4x4", is_slippery=True)

# Obtaining environment details
print('Number of Actions:', env.action_space.n)
print('Number of States: ', env.observation_space.n)
```

```python
# 1. Policy iteration

gamma = 0.9 # discount factor 
theta = 0.001 # desired convergence error
opt_pol, opt_val, val_hist = pol_iter(env, gamma, theta) # perform value iterations
print("Optimal policy: ", opt_pol)
```

```python
# 2. Plot showing convergence of policy iteration

plt.plot(val_hist) # plot value history to show convergence
plt.title("Convergence of policy iteration")
plt.xlabel("No. of iteration")
plt.ylabel("Value error")
```

```python
# 3. Visualize optimal value and optimal policy

fancy_visual(opt_val, opt_pol)
```

```python
# 5. Find Q table

opt_q_fun = find_opt_q(env, gamma, opt_val) # compute  Q values for obtained optimal value
print(opt_q_fun.round(3))
```
